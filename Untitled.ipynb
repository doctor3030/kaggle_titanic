### Preparing data for classification for missing values of 'Embarked'.

```python
#preparing data for classification for missing values of 'Embarked'.
df = data_train.drop(columns=['Survived', 'Age'])
# df1 = df.loc[:, ['Pclass', 'Sex', 'SibSp', '', '', '']]

#removing lines with missing values
df = df.dropna()
# df

#encoding categorical variables
mapping_sex = {'male': -1, 'female': 1}
df.loc[:,['Sex']] = [mapping_sex[item] for item in df.Sex]

mapping_port = {'C':0, 'Q':1, 'S':2}
df.loc[:,['Embarked']] = [mapping_port[item] for item in df.Embarked]
```

```python
#deriving a dictionary of families (key = 'Passanger id', value = 'dataframe of people having same 'Ticket')
familys = {}
#looping through data
for i in df.index:
    #i-th ticket id
    ticket_i = df.loc[i,['Ticket']].values[0]
    #i-th  family
    family = df[df['Ticket']==ticket_i]
    
    #if family contains >=1 people than recalculate ticket fare for every family member: fare = 'Fare' / n-people
    if len(family.index)>1:
        familys[i] = family
        family.loc[:,['Fare']] = family['Fare'] / len(family.index)

#assing recalculated fares to families members in data set
for i in familys:
    for j in familys[i].index:
        df.loc[j,['Fare']] = familys[i].loc[j,['Fare']].values[0]

familys.keys()
```

```python
def class_hists(ax, data,
                column,
                target,
                bins="auto", 
                legend=False,
                scale_separately=True,
               dropna=True):
    """Grouped univariate histograms.

    Parameters
    ----------
    data : pandas DataFrame
        Input data to plot
    column : column specifier
        Column in the data to compute histograms over (must be continuous).
    target : column specifier
        Target column in data, must be categorical.
    bins : string, int or array-like
        Number of bins, 'auto' or bin edges. Passed to np.histogram_bin_edges.
        We always show at least 5 bins for now.
    ax : matplotlib axes
        Axes to plot into
    legend : boolean, default=False
        Whether to create a legend.
    scale_separately : boolean, default=True
        Whether to scale each class separately.

    Examples
    --------
    >>> from dabl.datasets import load_adult
    >>> data = load_adult()
    >>> class_hists(data, "age", "gender", legend=True)
    <matplotlib...
    """
    col_data = data[column].dropna()

    if ax is None:
        ax = plt.gca()
    if col_data.nunique() > 10:
        ordinal = False
        # histograms
        bin_edges = np.histogram_bin_edges(col_data, bins=bins)
        if len(bin_edges) > 30:
            bin_edges = np.histogram_bin_edges(col_data, bins=30)

        counts = {}
        for name, group in data.groupby(target)[column]:
            this_counts, _ = np.histogram(group, bins=bin_edges)
            counts[name] = this_counts
        counts = pd.DataFrame(counts)
    else:
        ordinal = True
        # ordinal data, count distinct values
        counts = data.groupby(target)[column].value_counts().unstack(target)
    if scale_separately:
        # normalize by maximum
        counts = counts / counts.max()
    bottom = counts.max().max() * 1.1
    for i, name in enumerate(counts.columns):
        if ordinal:
            ax.bar(range(counts.shape[0]), counts[name], width=.9,
                   bottom=bottom * i, tick_label=counts.index, linewidth=2,
                   edgecolor='k')
            xmin, xmax = 0 - .5, counts.shape[0] - .5
        else:
            ax.bar(bin_edges[:-1], counts[name], bottom=bottom * i, label=name,
                   align='edge', width=(bin_edges[1] - bin_edges[0]) * .9)
            xmin, xmax = bin_edges[0], bin_edges[-1]
        ax.hlines(bottom * i, xmin=xmin, xmax=xmax,
                  linewidth=1)
    if legend:
        ax.legend()
    ax.set_yticks(())
    ax.set_xlabel(column)
```

# Cross Validation

```python
# train test split
X = df.drop(columns=['Embarked', 'Ticket', 'PassengerId'])
# X = df['Fare']
y = df['Embarked']

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
# random_state=30, stratify=y)
```

```python
#Initializing MyLearner
mylearner = MyLearner()

#number of folds for cross validation
cv_folds = 5

#cross validation metrics
myscorer = MyScorer()
scoring = {'roc_auc_ovr':  metrics.make_scorer(myscorer.roc_auc_score_ovr, needs_proba=True),
           'roc_auc_ovo':  metrics.make_scorer(myscorer.roc_auc_score_ovo, needs_proba=True)}
```

```python
#initial model parameters
params = {'max_depth': 5,
          'n_estimators': 50,
          'subsample':0.8,
          'learning_rate': 0.1,
          'random_state': 30}

#instantiating model
gbm = GradientBoostingClassifier(**params)

#fitting model with initial parameters
results_title = gbm.__class__.__name__
mylearner.fit_model(gbm, X, y, 
                    scoring=scoring, cv_folds=cv_folds,
                    cv_report=True, class_report=False, feature_importance=True, conf_matrix=True, confm_norm=True, figure_title=results_title,
                    tune=False)
```

## Implementing Resampling

```python
strategy = 'not majority'
random_state = 30
knn = 5
n_jobs = -1

#using SMOTE
smote = imbOVS.SMOTE(sampling_strategy=strategy, random_state=random_state, k_neighbors=knn, n_jobs=n_jobs)

#using SMOTENC
cat_features = ['Sex', 'SibSp', 'Parch', 'Pclass']
cat_features = [list(X.columns).index(i) for i in cat_features]
smotenc = imbOVS.SMOTENC(categorical_features=cat_features, sampling_strategy=strategy, random_state=random_state, k_neighbors=knn, n_jobs=n_jobs)

#using KMeansSMOTE
kmsmote = imbOVS.KMeansSMOTE(sampling_strategy=strategy, random_state=random_state, k_neighbors=knn, n_jobs=n_jobs, 
                               kmeans_estimator=None, cluster_balance_threshold=0.1, density_exponent='auto')

#using SVMSMOTE
svmsmote = imbOVS.SVMSMOTE(sampling_strategy=strategy, random_state=random_state, k_neighbors=knn, n_jobs=n_jobs, m_neighbors=10, svm_estimator=None, out_step=0.5)

#using BorderlineSMOTE
blsmote = imbOVS.BorderlineSMOTE(sampling_strategy=strategy, random_state=random_state, k_neighbors=knn, n_jobs=n_jobs, m_neighbors=10, kind='borderline-1')

#using ADASYN
adasyn = imbOVS.ADASYN(sampling_strategy=strategy, random_state=random_state, n_neighbors=knn, n_jobs=n_jobs)

#defining resamplers
resamplers = [smote, smotenc, kmsmote, svmsmote, blsmote, adasyn]

for resampler in resamplers:
    X_res, y_res = resampler.fit_resample(X, y)
    results_title = '{0} \n Oversampling: {1}'.format(gbm.__class__.__name__, resampler.__class__.__name__)
    mylearner.fit_model(gbm, X_res, y_res, 
                      scoring=scoring, cv_folds=cv_folds,
                      cv_report=True, class_report=False, feature_importance=False, conf_matrix=True, confm_norm=True, figure_title=results_title,
                      tune=False)
```

## Best Resampler Results
