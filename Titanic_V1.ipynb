```python
#encoding Name
def name_encoder(data):
    data_encoded = []
    for i in data['Name']:
        lname0 = i.split(',')[0].strip() #last name
        ttl = i.split(',')[1].split('.')[0].strip() #title
        lname1 = (i.split('(')[-1].split(')')[0].split()[-1].strip() if '(' in i else '').replace('\"','') #second last name
        data_encoded.append([lname0, lname1, ttl])
    data_encoded = pd.DataFrame(data_encoded, columns=['lname0', 'lname1', 'ttl'], index=data.index)
    return pd.concat([data.reset_index(), data_encoded], axis=1).drop(columns=['index'])
```

```python
def title_encoder(data):
    mask = {'Don':'Mr',
            'Sir':'Mr',
            'Jonkheer':'Mr',
            'Mme':'Mrs',
            'Lady':'Mrs',
            'the Countess':'Mrs',
            'Dona':'Mrs',
            'Mlle':'Miss',
            'Ms':'Miss',
            'Major':'Mr',
            'Col':'Mr',
            'Capt':'Mr',
            'Rev':'Mr',
            'Dr':'Mr'}
    
    df = data['ttl'].apply(lambda i: mask[i] if i in mask else i)
    data = pd.concat([data.drop(columns=['ttl']), df], axis=1)
    if 797 in data['PassengerId'].values:
        data.loc[data_tot['PassengerId']==797, 'ttl'] = 'Mrs'
    mask = {'Mr':1, 'Mrs':2, 'Miss':3, 'Master':4}
    df = data['ttl'].apply(lambda i: mask[i])
    return pd.concat([data.drop(columns=['ttl']), df], axis=1)
```

```python
def grps_by_size(grp):
    #group size > 1
    grp1 = None
    #group size == 1
    grp0 = None
    for g in grp.groups:
        grp_ = grp.get_group(g)
        if len(grp.get_group(g).index)>1:
            if grp1 is None:
                grp1 = grp_
            else:
                grp1 = pd.concat([grp1, grp_])
        else:
            if grp0 is None:
                grp0 = grp_
            else:
                grp0 = pd.concat([grp0, grp_])
    return (grp0, grp1)
```

```python
#Encoding familie groups
def familie_groups_encoder(data, fam_grps):
    data['fgrp'] = 0
    for idx, grp in enumerate(fam_grps):
#         grp['fgrp'] = idx + 1
        for pid in grp['PassengerId']:
            data.loc[data['PassengerId']==pid, 'fgrp'] = idx + 1
    return data
```

```python
#Encode 'Fare'
def fare_encoder(data, encode_grps):
    for grp in encode_grps:
        fare = grp['Fare'].values[0] / len(grp.index)
        grp['Fare'] = fare
        for pid in grp['PassengerId']:
            if pid in data['PassengerId']:
                data.loc[data['PassengerId']==pid, 'Fare'] = fare
    if len(data.loc[data['Fare'].isnull()])>0:
        for pid in data.loc[data['Fare'].isnull(), 'PassengerId']:
            c = data.loc[data['PassengerId']==pid, 'Pclass'].values[0]
            data.loc[data['PassengerId']==pid, 'Fare'] = np.mean(data.loc[data['Pclass']==c, 'Fare'])
    return data
```

```python
#Encoding 'Sex'
def sex_encoder(data):
    mask = {'male':1, 'female':2}
    df = data['Sex'].apply(lambda i: mask[i])
    return pd.concat([data.drop(columns=['Sex']), df], axis=1)
```

```python
#Adding 'Familie size'
def fam_size(data):
    data['famsize'] = data['SibSp'] + data['Parch'] + 1
    return data
```

```python
def param_tune(model, X_train, y_train, param_ranges, scoring, cv, refit=''):
    if refit == '': refit = list(scoring.keys())[0]
    gsearch = GridSearchCV(estimator=model, param_grid=param_ranges, scoring=scoring, cv=cv, refit=refit,
                           n_jobs=-1)
    gsearch.fit(X_train, y_train)
    cv_res = pd.DataFrame(gsearch.cv_results_)
    cv_res_sel = []
    for i in list(scoring.keys()):
        cv_res_sel.append(
            cv_res.loc[cv_res['rank_test_%s' % i] == 1, ['mean_test_%s' % i, 'std_test_%s' % i]].values[0])

    #                 print(cv_res_sel)
    cv_res_sel = pd.DataFrame(cv_res_sel, index=list(scoring.keys()), columns=['mean', 'std'])
    cv_results = cv_res_sel.T
    cv_results = cv_results.apply(lambda i: np.absolute(i))
    return gsearch.best_estimator_, gsearch.best_params_, gsearch.best_score_, cv_results
```

```python
def cv_rep(model, X, y, cv, scoring):
    cv_res = model_selection.cross_validate(model, X, y, scoring=scoring, cv=cv, n_jobs=-1)
    cv_res = [[np.mean(cv_res['test_%s' % i]), np.std(cv_res['test_%s' % i])] for i in list(scoring.keys())]
    cv_results = pd.DataFrame(cv_res, index=list(scoring.keys()), columns=['mean', 'std'])
    cv_results = cv_results.T
    cv_results = cv_results.apply(lambda i: np.absolute(i))
    return cv_results
```

```python
def age_predictor(base_models, base_models_params, final_model, final_model_params, X, y, test_size, random_state, cv, scoring, refit):
    #Train/Test split
    X_train, X_test, y_train, y_test =model_selection.train_test_split(X, y, test_size=test_size, random_state=random_state)
    #Training base models
    best_base_models = []
    print('Base models score:')
    for model, param_ranges in zip(base_models, base_models_params):
        model.fit(X_train, y_train)
        best_model, best_params, best_score, cv_res = param_tune(model, X_train, y_train, param_ranges, scoring, cv, refit=refit)
        print('{0} \n {1} \n Best params: {2} \n'.format(model.__class__.__name__, cv_res, best_params))
        best_base_models.append(best_model)
    #Training final model
    X_train = add_stack_models_predictions(models=best_base_models, X=X_train)
    final_model.fit(X_train, y_train)
    print('Final model score: {0}'.format(final_model.__class__.__name__))
    for i, param_ranges in enumerate(final_model_params):
        print('Param set #{0}'.format(i+1))
        final_model, best_params, best_score, cv_res = param_tune(final_model, X_train, y_train, param_ranges, scoring, cv, refit=refit)
        print('{0} \n Best params: {1} \n'.format(cv_res, best_params))
    #Fitting test set
    X_test = add_stack_models_predictions(models=best_base_models, X=X_test)
    print('Test set score: \n {0}'.format(cv_rep(final_model, X_test, y_test, cv, scoring)))
    return final_model
```

```python
def model_optimal_features(model, X_train, y_train, candidates_todrop, param_ranges, scoring, refit):
    models = []
    #score for all features included
    opt_model = param_tune(model, X_train, y_train, param_ranges, scoring, cv, refit)
#     models.append(param_tune(model, X_train_, y_train, param_ranges, scoring, cv, refit=scoring))
    #create a list of columns to drop-test
    candidates = [[i] for i in candidates_todrop]
#     candidates = [[i] for i in X_train.columns]
    for cols in itertools.combinations(candidates, 2):
        candidates.append([cols[0][0], cols[1][0]])
    #train models dropping candidates
    for cols in candidates:
        X_train_ = X_train.drop(columns=cols)
        models.append(param_tune(model, X_train_, y_train, param_ranges, scoring, cv, refit))
    scores_ = [m[2] for m in models]
    scores_max = np.max(scores_)
    #optimal model
    cols_todrop = None
    if opt_model[2] < scores_max:
        cols_todrop = candidates[scores_.index(scores_max)]
        opt_model = models[scores_.index(scores_max)]
    return opt_model, cols_todrop
```

```python
#Age encoder
def age_encoder(data, model_age):
    
    data_na = data[data['Age'].isnull()]
    data_notna = data[data['Age'].notna()]
    #transform non-null age values
    pt = prepc.PowerTransformer()
    data_pt = pt.fit_transform(np.array(data_notna['Age']).reshape(-1,1))
    data_pt = pd.Series(data_pt.reshape(len(data_pt)), name='Age').to_frame()
    data_notna = pd.concat([data_notna.reset_index().drop(columns=['Age', 'index']), data_pt], axis=1)
    #predict missing values
    X = add_stack_models_predictions(models=[('abr',abr), ('bgr',bgr), ('etr',etr), ('rfr',rfr)], 
                                     X=data_na[['Pclass', 'Sex', 'ttl', 'fgrp', 'famsize']])
    y = pd.Series(model_age.predict(X), name='Age').to_frame()
    
    data_na = pd.concat([data_na.reset_index().drop(columns=['Age', 'index']), y], axis=1)
    return pd.concat([data_na, data_notna])
```
